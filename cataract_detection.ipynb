```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python
image_size = 224
```

```python
new_image = cv2.imread("./datasets/dissease eye recognition/preprocessed_images/5_left.jpg", cv2.IMREAD_COLOR)
new_image = cv2.resize(new_image, (image_size, image_size))
new_image = np.array(new_image).reshape(1, image_size, image_size, 3)
```

```python
saved_model = load_model("./eyedisease.h5")
```

```python
y_pred = saved_model.predict(new_image)
# classes_x=np.argmax(y_pred,axis=1)
prediction=np.where(y_pred > 0.5, 1, 0)
```

```python
plt.figure(figsize=(12,7))

pred_category = prediction[0]
     
if pred_category== 0:
    pred_label = "Normal"
else:
    pred_label = "Cataract"   
plt.subplot(2,5,1)
plt.imshow(new_image[0])
plt.xlabel("Prediction:{}".format(pred_label))
```
